```python
import pandas as pd
messages = pd.read_csv(
    r'E:\GenAI\4 - ML\SMSSpamCollection.txt', # make it raw string as there is 4 in the file name and python doesnt accept that
    sep='\t',
    names=["label", "message"]
)
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    # messages['message'][i] means: Get the message in the i‑th row of the "message" column of the DataFrame.
    review=re.sub('[^a-zA-z]',' ',messages['message'][i]) #This removes everything except letters (a–z or A–Z) from the message.
    review=review.lower()
    review=review.split()
    print(review)
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create Bag Of Words

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True) #take top 100 words with max freq : max_features
```

```python
X=cv.fit_transform(corpus).toarray()
```
